````python
from collections import namedtuple
import yt_dlp
import sys
import os.path
````

````python
TrimWindow = namedtuple("TrimWindow", "start end")
````

```
YouTube ID: ---1_cCGK4M  
Trim Window: (0.0, 10.0)  
Labels: [' "/m/01g50p', '/m/0284vy3', '/m/06d_3', '/m/07jdr', '/m/07rwm0c"']  
```

````python
video_id = "---1_cCGK4M"
window = TrimWindow(4.0, 8.0)
labels = ["/m/01g50p", "/m/0284vy3", "/m/06d_3", "/m/07jdr", "/m/07rwm0c"]
````

````python
duration = window.end - window.start
ydl_options = {
    "format": "flac/bestaudio/best",
    "postprocessors": [{
        "key": "FFmpegExtractAudio",
        "preferredcodec": "flac"
    }],
    "postprocessor_args": ["-ss", str(window.start), "-t", str(duration)],
    "paths": { "home": "data/" },
    "ffmpeg_location": "/usr/bin/ffmpeg"
}
with yt_dlp.YoutubeDL(ydl_options) as ydl:
    error_code = ydl.download(video_id)
````
